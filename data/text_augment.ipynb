```python
import pandas as pd
import json
import os
```

```python
path='./group-object/'
label=path+'modified_object_name.json'
mapping=path+'mapping_references.csv'
```

```python
with open(label, 'r') as j:
    dict= json.load(j)
    label_df = pd.DataFrame.from_dict(dict,orient='index',columns=['label'])
    label_df.index = label_df.index.astype(int) 
label_df
```

```python
map_df=pd.read_csv(mapping).set_index('mappingID')
map_df
```

```python
#join_df=pd.merge(map_df,label_df,left_index=True, right_index=True)
join_df=map_df.join(label_df).set_index('ID')
join_df
```

```python
#join_df.to_csv('new_label.csv',header=None,index=None)
```

```python
aug_list={}
origin=join_df.index.to_list()
group=path+'data/'
for folder in origin:
    if not os.path.exists(group+folder): continue
    aug_list[folder]=[]
    for file in os.listdir(group+folder):
        if file[:-4]!=folder:
            aug_list[folder].append(file[:-4])
aug_list
```

```python
s=0
for folder in aug_list:
    for file in aug_list[folder]: 
        try:
            join_df.at[file,'label']=join_df.at[folder,'label']
            s+=1
        except:
            pass
join_df=join_df.dropna()
join_df
#print(s)
```

```python
join_df=pd.read_csv(path+'new_train_tex.csv')
join_df
```

```python
file_col=[i+'.obj' for i in join_df['obj_id'].to_list()]
join_df['obj_filename']=file_col
join_df
```

```python
text_id={}
count=1
name_set=set()
for i in join_df['tex'].to_list():
    name_set.add(i)
for i in name_set:
    text_id[i]='a'+str(count)
    count+=1
text_id
```

```python
textid_df=pd.DataFrame.from_dict(text_id,orient='index',columns=['text_id'])
textid_df.loc['elephant','text_id']
    
```

```python
join_df=join_df.join(textid_df,on='tex')
join_df = join_df.reset_index()
join_df
```

```python
#join_df=join_df.drop(['index'],axis=1)
join_df = join_df[[ 'obj_id', 'text_id', 'obj_filename', 'tex']]
join_df
```

```python
test_path=path+'test_tex.csv'
test_df=pd.read_csv(test_path)
test_list=test_df['obj_id'].to_list()
test_df
```

```python
join_df=join_df[~join_df['obj_id'].isin(test_list)]
join_df
```

```python
train_path='./csv/train_tex.csv'
train_df=pd.read_csv(train_path)
train_list=train_df['obj_id'].to_list()
train_df
```

```python
join_df=join_df[~join_df['obj_id'].isin(train_list)]
join_df
```

```python
join_df=join_df[~join_df['tex'].str.contains('_bird')]
join_df
```

```python
join_df.to_csv('./group-object/labeled_data.csv',header=None,index=False)
```

# K Fold

```python
path='./kfold/'
df_full=pd.read_csv(path+'full_tex.csv')
df_full
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5, shuffle = False, random_state = None)

fold = 0
for i, (train_index, test_index) in enumerate(kf.split(df_full)):
    train = df_full.iloc[train_index]
    val = df_full.iloc[test_index]
    train.to_csv(path+f'train_tex_{fold}.csv', index = False)
    val.to_csv(path+f'test_tex_{fold}.csv', index = False)
    fold += 1
kf.split(df_full)
```

```python
for fold in range(5):
    print('Test fold', fold)
    df_train_0 = pd.read_csv(path+f'train_tex_{fold}.csv')
    df_val_0 = pd.read_csv(path+f'test_tex_{fold}.csv')
    print('Train:', len(df_train_0))
    print('Val:', len(df_val_0))
```
